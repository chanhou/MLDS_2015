```python
import numpy as np
```

```python
collection_data = []
collection_ppl = []

with open('../hw1/MLDS_HW1_RELEASE_v1/fbank/train.ark') as f:
    
#     count = 0
    for lines in f :
        frames = lines.split(' ')
#         print frames
#         break
        frame2float = np.array([ float(x) for x in frames[1:] ])
        
        collection_data.append(frame2float)
        collection_ppl.append(frames[0])
```

```python
collection_data = np.array(collection_data)
collection_ppl = np.array(collection_ppl)
```

```python
from sklearn import preprocessing

collection_data
```

```python
X_scaled = preprocessing.scale(collection_data)

X_scaled
```

```python
f = open("hwb_normalize.dat", 'w+')
# f.write("id,feature\n")
for ppl, data in zip(collection_ppl, X_scaled):
    f.write("{0} ".format(ppl))
    # for ddd in data:
    for i in range (len(data)):
        if( (i+1)!= len(data)):
            f.write("{0} ".format(data[i]))
        else:
            f.write("{0}\n".format(data[i]))
#     f.write("\n")
#     break
f.close()
```
